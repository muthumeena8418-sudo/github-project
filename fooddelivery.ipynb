```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample Food Delivery Dataset
data = {
    "OrderID": [101,102,103,104,105,106,107,108,109,110],
    "Time": [18,19,20,21,20,22,21,20,19,18],
    "Location": ["Chennai","Madurai","Chennai","Coimbatore",
                 "Madurai","Chennai","Madurai","Chennai",
                 "Coimbatore","Madurai"],
    "Item": ["Pizza","Burger","Pizza","Fried Rice",
             "Pizza","Burger","Pizza","Pizza",
             "Burger","Pizza"],
    "Amount": [250,180,300,220,280,190,310,270,200,290]
}

df = pd.DataFrame(data)

# Dataset Preview Screenshot
print("Dataset Preview")
print(df.head())

# Cleaned Dataset Screenshot
df = df.drop_duplicates()
print("\nCleaned Dataset")
print(df)

# Bar Chart - Most Popular Dishes
plt.figure(figsize=(6,4))
df["Item"].value_counts().plot(kind="bar")
plt.title("Most Popular Dishes")
plt.xlabel("Food Item")
plt.ylabel("Orders")
plt.show()

# Line Chart - Peak Order Times
plt.figure(figsize=(6,4))
df["Time"].value_counts().sort_index().plot(kind="line", marker="o")
plt.title("Peak Order Times")
plt.xlabel("Hour")
plt.ylabel("Number of Orders")
plt.show()

# Pie Chart - Orders by Location
plt.figure(figsize=(6,4))
df["Location"].value_counts().plot(kind="pie", autopct="%1.1f%%")
plt.title("Orders by Location")
plt.ylabel("")
plt.show()
```
